```python
import numpy as np
from results_L5 import results as results_5    
from results_L6 import results as results_6    
from results_L7 import results as results_7    
from results_L8 import results as results_8 
from results_L5_1 import results as results_5_1
from results_L4 import results as results_4 
from results_L5_2 import results as results_5_2

from matplotlib.pyplot import *
```

```python
results = [results_4, results_5_2, results_6, results_7, results_8]
maxiter = [450_000, 600_000, 45_000, 13_000, 1_600]
```

```python
for i in range(5):
    interesting = results[i]
    x = range(1,2*(i+4)+2)
    plot(x, [interesting[k][0][0] for k in range(len(interesting))], "-o", label="Mean convergence")
    plot(x, [interesting[k][0][1] for k in range(len(interesting))], "-o", label="Var convergence")
    legend()
    xlabel('k, computational step')
    title('montecarlo_convergence, L = ' + str(i+5))
    xticks(x)
    show()
```

```python
for i in range(5):
    interesting = results[i]
    x = range(1,2*(i+4)+2)
    plot(x, [len(interesting[k][1]) for k in range(len(interesting))], "-o", label="Iterations completed")
    legend()
    xlabel('k, computational step')
    ylabel('iterations')
    title('Montecarlo iterations, L = '+ str(i+5)+ ', maxiter = ' + str(maxiter[i]))
    xticks(x)
    #savefig("iteration_montecarlo_convergence_L="+str(i+5), dpi=300)    
    show()
```

## Estimate convergence

```python
log_iter = np.log2([len(results[1][k][1]) for k in range(len(results[1]))])
plot(range(len(log_iter)), log_iter, 'o')
m = np.mean(np.diff(log_iter[:7]))
q = log_iter[0]
x = np.array(range(11))
plot(x, m * x + q)
title("Steps for convergence, log scale")
xlabel("k")
ylabel("log2(iterations)")
savefig("iterations_convergence_L_5.png", dpi=300)
#print(2 **(m * 10 + q))
```
