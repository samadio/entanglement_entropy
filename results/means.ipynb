## QUESTION: Can average entropy decrease when k increase?

Since you have k+L qubits, you'd expect entropy to increase, in general, for same N when k increase, or at worst be the same. Can it decrease?

```python
N = numbers[idx]
res = results[idx]
r_L = r_list[idx]

max_k_plotted = 2*L+1

last_means = list(map(lambda x: x[max_k_plotted-1], means_stratified))
_, ordered_N = zip(*sorted(zip(last_means, N),reverse=True))
order_of_N = [N.index(ordered_N[n]) for n in range(len(N))]

print([N[n] for n in order_of_N])

lines_per_fig = 9
for i in range(len(N)//lines_per_fig+1):
    for n in order_of_N[lines_per_fig*i:lines_per_fig*(i+1)]:
        plt.plot(range(1,2*L+2)[:max_k_plotted], means_stratified[n][:max_k_plotted], label= "N=" + str(N[n])+", r="+str(r_L[n]))

    plt.grid(b=True,linestyle='--', dashes = (0,0,1,3))
    plt.title("L = " + str(L))
    plt.xlabel("k")
    plt.ylabel("$\mu_S$")   
    lgd = plt.legend(bbox_to_anchor=(1.0, 1.0))
    plt.tight_layout()
    #plt.savefig("S_across_k_L_"+str(L)+"_WithIQFT_"+str(i+1), dpi=400, bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.show()
```

## Answer: Yes, but not really
We simulated all the bipartitions here, so yes, mean entropy can decrease, but not significally

## Sigmoid fitting no IQFT for each N and see if something reminds of N, r or divisors

```python
import scipy.optimize as opt

def f(x, a, b, c, d):
    return a / (1. + np.exp(-c * (x - d))) + b

x = range(1,2*L+1)
for n in range(len(N)):
    y = means_stratified[n][:-1]
    (a_, b_, c_, d_), _ = opt.curve_fit(f, x, y, p0=[y[-1], 1,1,1])
    params = (a_, b_, c_, d_)
    y_fit = f(x, a_, b_, c_, d_)
    fig, ax = plt.subplots(1, 1, figsize=(6, 4))
    ax.plot(x, y, 'd')
    ax.plot(x, y_fit, '-')
    ax.set_title('Sigmoid fit, N='+str(N[n])+', r='+str(r_L[n]))
    (a_, b_, c_, d_) = (str(a_), str(b_), str(c_), str(d_))
    string = str(b_)[:4] + "+ \\frac{{{}}}{{1+e^{{-{}(k-{})}}}}".format(a_[:4],c_[:4],d_[:4])
    ax.text(1, 1.4, "S(k) = $"+string+"$", style='normal', fontsize=20, 
    bbox={'facecolor': 'grey', 'alpha': 0.5, 'pad': 10})
    
    plt.xlabel("k")
    plt.ylabel("$\mu_S$")
    plt.tight_layout()
    
    if np.abs(y[-1] - y[3])>0.5:
        print(N[n])
        #plt.savefig("Sigmoid_fit_N_"+str(N[n]), dpi=300)
    plt.show()
```
