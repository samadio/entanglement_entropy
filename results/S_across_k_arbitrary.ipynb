```python
from scipy.special import binom
from numpy import floor,ceil
from numba import njit, prange

@njit(nogil=True)
def entropy_value(l,r,in_control,control_interesting=None, control_size = None):
    '''
    returns max_entr * n_edges1 / n_edges2
    with max_entr: log2(r), n_edges1 = n_important edges between control and target outside bipartition
    n_edges2 = n_import edges between control and target
    '''
    exact = np.log2(r)
    total_interesting = int(ceil(exact))
    if control_interesting == None: control_interesting = in_control
    if control_size == None: control_size = total_interesting
    num_qubits = l + control_size
    bip_size = num_qubits // 2
    total_connections = l * total_interesting
    internal = control_interesting * (bip_size - in_control)
    if internal ==0: return exact
    external_connections = total_connections - internal
    return exact / total_connections * external_connections        


#theoretical mu at k=ceil(log_2r) for k+l qubits considering no entropy in target register,
#entropy averaged on all bipartitions and maximally multipartite
def theoretical_mu_no_target_entropy(r,l,k=None):
    exact = np.log2(r)
    k_convergence = int(ceil(exact))
    if k==None: k = k_convergence
    if k < k_convergence:
        return -1
    numb_qubits = k + l
    bigger_register_size = max(l, k)
    smaller_register_size = numb_qubits - bigger_register_size
    bipartition_size = numb_qubits // 2
    total_bipartitions = binom(numb_qubits, bipartition_size)
    
                                #numb of ways to choose the remaining qubits from big register 
    if k == None or k ==k_convergence:
        mu = np.sum([entropy_value(l,r,i) * binom(bigger_register_size,bipartition_size - i) 
        #numb of ways to select i qubits from small register          
        * binom(smaller_register_size,i) for i in range(0,bipartition_size+1)])
        return mu / total_bipartitions
    else: #smaller register = target register
        mu = np.zeros(k_convergence + 1)
        #number of interesting control qubits
        for j in range(0,k_convergence+1):
            mu[j] = np.sum([entropy_value(l,r,in_control,j,k) *
               binom(k-k_convergence,in_control-j) * binom(k_convergence,j)*
               binom(l, bipartition_size - in_control)
               for in_control in range(j,bipartition_size+1)])
        return np.sum(mu) / total_bipartitions

    
def theoretical_no_target_entropy(r,l,k=None):
    exact = np.log2(r)
    k_convergence = int(ceil(exact))
    if k==None: k = k_convergence
    if k < k_convergence:
        return -1
    numb_qubits = k + l
    bigger_register_size = max(l, k)
    smaller_register_size = numb_qubits - bigger_register_size
    bipartition_size = numb_qubits // 2
    total_bipartitions = int(binom(numb_qubits, bipartition_size))
    
    mu=[]
                                #numb of ways to choose the remaining qubits from big register 
    if k == None or k ==k_convergence:
        new_element = [[entropy_value(l,r,i)] * int(binom(bigger_register_size,bipartition_size - i) 
        #numb of ways to select i qubits from small register          
        * binom(smaller_register_size,i)) for i in range(0,smaller_register_size+1)]
        
        new_element = [item for sublist in new_element for item in sublist]
        return np.array(new_element)
    
    else: #smaller register = target register
        mu = []
        #number of interesting control qubits
        for j in range(0,k_convergence+1):
            new_element = [[entropy_value(l,r,in_control,j,k)] *
               int(binom(k-k_convergence,in_control-j) * binom(k_convergence,j)*
               binom(l, bipartition_size - in_control))
               for in_control in range(j,k+1)]
            new_element = [item for sublist in new_element for item in sublist]
            mu += new_element
        return np.array(mu)
    
max_k = 12
k_conv = 4
model_16 = [theoretical_no_target_entropy(16, 6,k) for k in range(4,max_k+1)]
```
